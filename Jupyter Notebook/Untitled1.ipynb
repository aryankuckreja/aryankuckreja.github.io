```python
import matplotlib.pyplot as plt
import numpy as np

plt.title("Stem Cell Transplants In Canada")
plt.ylabel("Amount of Stem Cell Transplants")
plt.xlabel("Years")


x1 = np.array([1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020])
y1 = np.array([700, 1800, 3000, 4000, 4800, 6400, 8000, 7000])
x2 = np.array([1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020])
y2 = np.array([0, 2900, 8000, 7000, 8000, 10000, 14000, 12000])

plt.plot(x1, y1, label = "Allogenic", color="orange")
plt.plot(x2, y2, label = "Autologous", color="blue")

plt.legend(loc="upper left")

plt.plot(x1, y1, linestyle='-', marker='o', color="black", label='line with marker')
plt.plot(x2, y2, linestyle='-', marker='o', color='black', label='line with marker')



plt.grid(axis='y', linestyle='-')
plt.grid(axis='x', linestyle='-')



plt.plot(x1, y1, x2, y2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

labels = ["Organ Toxicity (9%)", "IPn (2%)", "Infection (16%)", "Other (19%)", "Relapse (42%)", "GVHD (12%)"]
sizes = [9, 2, 16, 19, 42, 12]
colors = ["b", "orange", "g", "r", "#A020F0", "#680200"]

patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90)
plt.legend(labels,  loc="best")

plt.axis('equal')
plt.title("Causes of Death After Stem Cell Transplantation")

fracs = [9, 2, 16, 19, 42, 12]
total = sum(fracs)
explode = (0, 0, 0, 0, 0, 0)

plt.pie(fracs, explode=explode, labels=labels,
   autopct=lambda p: '{:.0f}%'.format(p * total / 100),
   shadow=True, startangle=90)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
men = [12, 30, 1, 8, 22]
women = [28, 6, 16, 5, 10]
br1 = np.arange(len(men))
br2 = [x + barWidth for x in br1]
plt.grid(axis='y', linestyle='-')
 
plt.bar(br1, men, color ='b', width = barWidth,
        edgecolor ='grey', label ='MEN')
plt.bar(br2, women, color ='m', width = barWidth,
        edgecolor ='grey', label ='WOMEN')
 
plt.xlabel('Ages', fontweight ='bold', fontsize = 15)
plt.ylabel('Net Survival (%)', fontweight ='bold', fontsize = 15)
plt.title("Leukaemia Survival Statistics", fontweight ='bold', fontsize = 15 )
plt.xticks([r + barWidth for r in range(len(men))],
        ['15-39', '40-49', '50-59', '60-69', '70-79'])
 
plt.legend()
plt.show()
```
